```python
# create_sample_sales_data.py
import pandas as pd
import numpy as np
import os
from datetime import datetime, timedelta

# Create folder if not exists
os.makedirs("data/sales_reports", exist_ok=True)

# Date range - last 30 days
dates = pd.date_range(end=datetime.today(), periods=30)

# Sample products
products = [
    "Hayati Pro Max+ 6K [Blueberry Pomegranate / 20mg]",
    "Elf Bar 600 [Peach Ice / 20mg]",
    "Oxva Xlim Pods [Strawberry Kiwi / 10mg]",
    "Lost Mary BM600 [Grape Ice / 20mg]",
    "Hayati Remix 2400 [Watermelon Bubblegum / 10mg]"
]

# Generate sales for each day
records = []
for date in dates:
    for product in products:
        base_sales = np.random.randint(20, 80)
        fluctuation = np.random.randint(-10, 10)
        sales = max(5, base_sales + fluctuation)
        records.append({
            "Date": date.strftime("%Y-%m-%d"),
            "Product": product,
            "Sales_Units": sales
        })

df = pd.DataFrame(records)

# Save daily sales files (one per date)
for d, daily_df in df.groupby("Date"):
    daily_df.to_csv(f"data/sales_reports/sales_{d}.csv", index=False)

print(f"✅ Created {len(df)} sales records across {len(products)} products for 30 days.")
```
